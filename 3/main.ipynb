```python
import numpy as np

np.random.seed(0)

X_size = 1000
X = np.random.exponential(scale=1.0, size=X_size)
```

```python
X
```

```python
def MAD(Xd):
    median = np.median(Xd)
    absolute_deviations = np.abs(Xd - median)
    
    return np.median(absolute_deviations)
```

```python
REAL_MAD_X_VALUE = MAD(X)
REAL_MAD_X_VALUE
```

```python
# BOOTSTRAP
M = 100000

bootstrap_xs = np.array(
    [np.random.choice(X, len(X), replace=True) for _ in range(M)]
)

bootstrap_xs[0]
```

```python
bootstrap_x_mads = [MAD(bootstrap_x) for bootstrap_x in bootstrap_xs]

bootstrap_x_mads.sort()

bootstrap_x_mads
```

```python
P = 95

e = int(M * (P/2) // 100)

cd_x = bootstrap_x_mads[e:M-e]
cd_x
```

```python
print(f"  ({P}%): {cd_x[0]} - {cd_x[-1]}.")
print(f" : {REAL_MAD_X_VALUE}")
```
